```python
import geopandas as gpd
import pandas as pd
import shared
import osmnx
%matplotlib inline
```

```python
gdf = gpd.read_geocsv("cache/split_parcels_unioned.csv")
```

```python
intersections = gpd.sjoin(gdf.copy(), gdf.copy())
```

```python
overlap_area = shared.compute_overlap_areas(intersections, gdf)
```

```python
intersections["overlap_area"] = overlap_area
```

```python
large_intersections = intersections[(intersections.overlap_area > 100) & (intersections.apn_left != intersections.apn_right)]
large_intersections
```
